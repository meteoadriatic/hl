```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
zagreb = pd.read_csv('Zagreb.csv', parse_dates=['Date'])
rijeka = pd.read_csv('Rijeka.csv', parse_dates=['Date'])
osijek = pd.read_csv('Osijek.csv', parse_dates=['Date'])
split = pd.read_csv('Split.csv', parse_dates=['Date'])
```

```python
zagreb['mean'] = (zagreb['06z'] + zagreb['18z']) / 2
rijeka['mean'] = (rijeka['06z'] + rijeka['18z']) / 2
osijek['mean'] = (osijek['06z'] + osijek['18z']) / 2
split['mean'] = (split['06z'] + split['18z']) / 2

zagreb['error_18z'] = zagreb['18z'] - zagreb['Actual']
zagreb['error_06z'] = zagreb['06z'] - zagreb['Actual']
zagreb['error_mean'] = zagreb['mean'] - zagreb['Actual']
rijeka['error_18z'] = rijeka['18z'] - rijeka['Actual']
rijeka['error_06z'] = rijeka['06z'] - rijeka['Actual']
rijeka['error_mean'] = rijeka['mean'] - rijeka['Actual']
osijek['error_18z'] = osijek['18z'] - osijek['Actual']
osijek['error_06z'] = osijek['06z'] - osijek['Actual']
osijek['error_mean'] = osijek['mean'] - osijek['Actual']
split['error_18z'] = split['18z'] - split['Actual']
split['error_06z'] = split['06z'] - split['Actual']
split['error_mean'] = split['mean'] - split['Actual']

zagreb_18z_bias = zagreb['error_18z'].mean().round(2)
zagreb_06z_bias = zagreb['error_06z'].mean().round(2)
zagreb_mean_bias = zagreb['error_mean'].mean().round(2)
rijeka_18z_bias = rijeka['error_18z'].mean().round(2)
rijeka_06z_bias = rijeka['error_06z'].mean().round(2)
rijeka_mean_bias = rijeka['error_mean'].mean().round(2)
osijek_18z_bias = osijek['error_18z'].mean().round(2)
osijek_06z_bias = osijek['error_06z'].mean().round(2)
osijek_mean_bias = osijek['error_mean'].mean().round(2)
split_18z_bias = split['error_18z'].mean().round(2)
split_06z_bias = split['error_06z'].mean().round(2)
split_mean_bias = split['error_mean'].mean().round(2)

zagreb_18z_mae = zagreb['error_18z'].abs().mean().round(2)
zagreb_06z_mae = zagreb['error_06z'].abs().mean().round(2)
zagreb_mean_mae = zagreb['error_mean'].abs().mean().round(2)
rijeka_18z_mae = rijeka['error_18z'].abs().mean().round(2)
rijeka_06z_mae = rijeka['error_06z'].abs().mean().round(2)
rijeka_mean_mae = rijeka['error_mean'].abs().mean().round(2)
osijek_18z_mae = osijek['error_18z'].abs().mean().round(2)
osijek_06z_mae = osijek['error_06z'].abs().mean().round(2)
osijek_mean_mae = osijek['error_mean'].abs().mean().round(2)
split_18z_mae = split['error_18z'].abs().mean().round(2)
split_06z_mae = split['error_06z'].abs().mean().round(2)
split_mean_mae = split['error_mean'].abs().mean().round(2)
```

```python
fig, ax = plt.subplots(figsize=(18, 14))

ax.plot(zagreb['Date'], zagreb['error_18z'], color='black', linestyle=':', label='ZG 18z, bias={}C'.format(zagreb_18z_bias))
ax.plot(zagreb['Date'], zagreb['error_06z'], color='black', linestyle='-', label='ZG 06z, bias={}C'.format(zagreb_06z_bias))
ax.plot(rijeka['Date'], rijeka['error_18z'], color='green', linestyle=':', label='RI 18z, bias={}C'.format(rijeka_18z_bias))
ax.plot(rijeka['Date'], rijeka['error_06z'], color='green', linestyle='-', label='RI 06z, bias={}C'.format(rijeka_06z_bias))
ax.plot(osijek['Date'], osijek['error_18z'], color='blue', linestyle=':', label='OS 18z, bias={}C'.format(osijek_18z_bias))
ax.plot(osijek['Date'], osijek['error_06z'], color='blue', linestyle='-', label='OS 06z, bias={}C'.format(osijek_06z_bias))
ax.plot(split['Date'], split['error_18z'], color='red', linestyle=':', label='ST 18z, bias={}C'.format(split_18z_bias))
ax.plot(split['Date'], split['error_06z'], color='red', linestyle='-', label='ST 06z, bias={}C'.format(split_06z_bias))
plt.axhline(y=0, color='grey', linestyle='--', lw=0.5)
plt.legend()
plt.title("BIAS")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 14))

ax.plot(zagreb['Date'], zagreb['error_mean'], color='black', linestyle='-', lw=2, label='ZG mean, bias={}C'.format(zagreb_mean_bias))
ax.plot(rijeka['Date'], rijeka['error_mean'], color='green', linestyle='-', lw=2, label='RI mean, bias={}C'.format(rijeka_mean_bias))
ax.plot(osijek['Date'], osijek['error_mean'], color='blue', linestyle='-', lw=2, label='OS mean, bias={}C'.format(osijek_mean_bias))
ax.plot(split['Date'], split['error_mean'], color='red', linestyle='-', lw=2, label='ST mean, bias={}C'.format(split_mean_bias))
plt.axhline(y=0, color='grey', linestyle='--', lw=0.5)
plt.legend()
plt.title("BIAS - mean temperature from 18z and 06z runs")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 14))

ax.plot(zagreb['Date'], zagreb['error_18z'].abs(), color='black', linestyle=':', label='ZG 18z, mae={}C'.format(zagreb_18z_mae))
ax.plot(zagreb['Date'], zagreb['error_06z'].abs(), color='black', linestyle='-', label='ZG 06z, mae={}C'.format(zagreb_06z_mae))
ax.plot(rijeka['Date'], rijeka['error_18z'].abs(), color='green', linestyle=':', label='RI 18z, mae={}C'.format(rijeka_18z_mae))
ax.plot(rijeka['Date'], rijeka['error_06z'].abs(), color='green', linestyle='-', label='RI 06z, mae={}C'.format(rijeka_06z_mae))
ax.plot(osijek['Date'], osijek['error_18z'].abs(), color='blue', linestyle=':', label='OS 18z, mae={}C'.format(osijek_18z_mae))
ax.plot(osijek['Date'], osijek['error_06z'].abs(), color='blue', linestyle='-', label='OS 06z, mae={}C'.format(osijek_06z_mae))
ax.plot(split['Date'], split['error_18z'].abs(), color='red', linestyle=':', label='ST 18z, mae={}C'.format(split_18z_mae))
ax.plot(split['Date'], split['error_06z'].abs(), color='red', linestyle='-', label='ST 06z, mae={}C'.format(split_06z_mae))
plt.axhline(y=0, color='grey', linestyle='--', lw=0.5)
plt.legend()
plt.title('MAE')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 14))

ax.plot(zagreb['Date'], zagreb['error_mean'].abs(), color='black', linestyle='-', lw=2, label='ZG mean, mae={}C'.format(zagreb_mean_mae))
ax.plot(rijeka['Date'], rijeka['error_mean'].abs(), color='green', linestyle='-', lw=2, label='RI mean, mae={}C'.format(rijeka_mean_mae))
ax.plot(osijek['Date'], osijek['error_mean'].abs(), color='blue', linestyle='-', lw=2, label='OS mean, mae={}C'.format(osijek_mean_mae))
ax.plot(split['Date'], split['error_mean'].abs(), color='red', linestyle='-', lw=2, label='ST mean, mae={}C'.format(split_mean_mae))
plt.axhline(y=0, color='grey', linestyle='--', lw=0.5)
plt.legend()
plt.title('MAE - mean temperature from 18z and 06 runs')
plt.show()
```
